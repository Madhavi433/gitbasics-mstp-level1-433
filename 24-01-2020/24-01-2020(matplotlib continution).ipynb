## creating subplot using numpy with sine

```python
import numpy as np
```

```python
t=np.arange(0.0,0.5,0.1)
s=np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s)
```

```python
t=np.arange(0.0,0.5,0.1)
s=np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s)
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
```

```python
t=np.arange(0.0,0.5,0.1)
s=np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,'rx')
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
```

```python
t=np.arange(0.0,0.5,0.1)
s=np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,'rx',label="graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
plt.show()
```

```python
#with nan values
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.exp(-t)*np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,label="1st graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
t[0]=np.nan
t[1]=np.nan
plt.subplot(2,1,2)
plt.plot(t,s,label="2nd graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
plt.legend()
plt.show()
```

## By using subplot to generate hist,scatter

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(222)
plt.plot(np.random.rand(30))
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(222)
plt.plot(np.random.rand(30))
ax1.hist(np.random.rand(50),bins=20,color='g')
```

## 3d Graph

```python
from mpl_toolkits.mplot3d import axes3d
```

from matplotlib import style

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
plt.show()
```

```python
x=[2,3,4,5]
y=[55,99,6,8]
z=[4,5,9,2]
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.plot(x,y,z)
plt.show()
plt.xlabel("x")
plt.ylabel("y")
#plt.zlabel("z")
plt.title("3d projection")
plt.show()
```

```python
x=[2,3,4,5]
y=[55,99,6,8]
z=[4,5,9,2]
dx=np.ones(4)
dy=np.zeros(4)
dz=np.ones(4)
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.bar3d(x,y,z,dx,dy,dz,color='m')
plt.xlabel("x")
plt.ylabel("y")
#plt.zlabel("z")
plt.title("3d projection")
plt.show()
```

```python
x=[2,3,4,5]
y=[55,99,6,8]
z=[4,5,9,2]
dx=np.ones(4)
dy=np.zeros(4)
dz=np.ones(4)
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.scatter(x,y,z,color='g')
plt.xlabel("x")
plt.ylabel("y")
#plt.zlabel("z")
plt.title("3d projection")
plt.show()
```
