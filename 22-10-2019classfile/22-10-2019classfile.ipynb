# Draw a plot for time and distance by using comprehensive 
# distance=2*time

```python
import matplotlib.pyplot as m
x=[i for i in range(1,10)]
y=[2*i for i in range(1,10)]
m.xlabel('time')
m.ylabel('distance')
m.title('lineplot')
m.plot(x,y)
m.show()
```

# Matplotlib ticks
## ticks is values within a coordinates

```python
import matplotlib.pyplot as pl
x=list(range(5,51,5))
print(x)
y=list(range(1,11))
print(y)
pl.plot(y,x)
pl.title('A Graph with a example to ticks')
pl.xlabel('Natural numbers')
pl.ylabel('Values with ++5 up to 50')
pl.show()
```

```python
import matplotlib.pyplot as pl
y=[5,10,15,20,25,30,35,40,45,50]
print(y)
x=[9,8,5,6,3,2,1,4,7,10]
print(x)
pl.plot(x,y)
pl.title('Matplotlib-Ticks')
pl.xlabel('10 Natural numbers')
pl.ylabel('5-50 values with 5+ upto 50')
pl.show()
```

```python
import matplotlib.pyplot as pl
import numpy as np
pl.xticks(np.arange(10,20,1))
pl.yticks(np.arange(5,51,5))
pl.plot(x,y)
print(x)
print(y)
pl.title('Matplotlib-Ticks')
pl.xlabel('10 Natural numbers')
pl.ylabel('5-50 values with 5+')
pl.show()
```

# Representing a sine wave
A=sine(wt)
A=Amplitude
w=Frequency
t=time

```python
import numpy as np
import matplotlib.pyplot as pl
Time=np.arange(1,10,0.1)
Amplitude=np.sin(Time)
pl.grid()
pl.plot(Time,Amplitude)
pl.title('sinewave')
pl.xlabel('Time')
pl.ylabel('Amplitude')
pl.show()
```

```python
import pandas as pd
df=pd.read_csv("income.csv")
df
```

```python
import matplotlib.pyplot as pl
years=df.columns[2::]
Alaska_income=df.iloc[1,2:]
pl.grid()
pl.plot(years,Alaska_income)
pl.title("income of Alaska from 2005 to 2013")
pl.xlabel("years")
pl.ylabel("Alaska_income")
pl.savefig("income of Alaska from 2005 to 2013.jpg")
pl.show()
```

# plot for all states income for 2005-2013
## x-axis=years
## y-axis=all States income

```python
import matplotlib.pyplot as pl
import pandas as pd
years=df.columns[-5::]
All_States_income=df.iloc[0:,2:]
pl.grid()
pl.plot(years,All_States_income)
pl.title("income of all states from 2005 to 2013")
pl.xlabel("years")
pl.ylabel("All_States_income")
pl.savefig("income of all states from 2005 to 2013.jpg")
pl.show()
```
