# Ridge Regression=sum of error+sum of sqres of magnitude of coefficient(alpha)

```python
from sklearn.linear_model import Ridge
rr=Ridge(alpha=1,solver='svd')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
rr.fit(x_train,y_train)
```

```python
y_predict=rr.predict(x_train)
y_predict
```

# score

```python
rr.score(x_train,y_train)
```

```python
rr.score(x_test,y_test)
```

```python
rr.coef_
```

```python
rr.intercept_
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin=LinearRegression()
lin.fit(x,y)
```

```python
lin.fit(x_train,y_train)
```

```python
lintrain_score=lin.score(x_train,y_train)
print("lin Train score",lintrain_score)
```

```python
lintest_score=lin.score(x_test,y_test)
print("lin Test score",lintest_score)
```

```python
lin.coef_
```

```python
import matplotlib.pyplot as plt
plt.plot(rr.coef_,marker="*",label="Ridge coefficient",color="r")
plt.plot(lin.coef_,marker=".",label="Linear coeffi",color="y")
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
plt.title("Ridge vs Linear")
plt.legend()
plt.show()
```

```python
rr1=Ridge(alpha=100,solver='svd')
```

```python
rr1.fit(x_train,y_train)
```

```python
rr1.coef_
```

```python
import matplotlib.pyplot as plt
plt.plot(rr1.coef_,marker="*",label="Ridge coefficient",color='g')
plt.plot(lin.coef_,marker=".",label="Linear coeff",color="y")
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
plt.title("Ridge vs Linear")
plt.legend()
plt.show()
```
